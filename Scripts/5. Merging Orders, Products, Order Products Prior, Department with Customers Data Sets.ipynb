```python
import pandas as pd
import numpy as np
import os
from os.path import join
import pickle
```

```python
path = '../Data'
```

```python
df_cus = pd.read_csv(join(path, 'Clean Data', 'clean_customers.csv'))
```

```python
df_main = pd.read_pickle(join(path, 'Clean Data', 'orders_prior_prods_dep.pkl'))
```

```python
df_cus.shape
```

```python
df_main.shape
```

```python
df_cus.head()
```

```python
# dropping Unnamed: 0 column
df_cus = df_cus.drop(columns = ['Unnamed: 0'])
```

```python
df_cus.head()
```

```python
pd.set_option('display.max_columns', None)
```

```python
# merging customers dataset with main dataset
df_main2 = df_main.merge(df_cus, on = 'user_id', how = 'inner', indicator = True)
```

```python
df_main2.head()
```

```python
# checking main2 results
value_counts = df_main2['_merge'].value_counts()
print(value_counts)
```

```python
# Checking for duplicate values
duplicate_rows = df_main2[df_main2.duplicated()]
if duplicate_rows.shape[0]==0:
    print('No Duplicates Found')
else:
    print(duplicate_rows.head())
```

```python
# checking for missing values
missing_values = df_main2.isnull().sum()
print(missing_values)
```

```python
# dropping the merge column
df_main2 = df_main2.drop(columns = ['_merge'])
```

```python
df_main2.head()
```

```python
del df_main
```

```python
del df_cus
```

```python
# saving main2 dataset
df_main2.to_pickle(join(path, 'Clean Data', 'main_dataset.pkl'))
```
