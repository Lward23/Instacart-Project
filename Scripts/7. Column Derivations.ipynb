```python
import pandas as pd
import numpy as np
import os
import pickle
from os.path import join
```

```python
path = '../Data'
```

```python
# Moving forward using the 70% sample of the final data set
df = pd.read_pickle(join(path, 'Clean Data', 'final_data_big.pkl'))
```

```python
df.shape
```

```python
df.head()
```

```python
# Looking into the order day of week
df['order_dow'].value_counts(dropna = False)
```

```python
# creating a new column for busiest days of week
def profiles(row):
    if row['order_dow'] == 0:
        return 'Busiest Day'
    elif row['order_dow'] == 4:
        return 'Least Busy Day'
    elif row['order_dow'] == 1:
        return '2nd Busiest Day'
    elif row['order_dow'] ==3:
        return '2nd Least Busy Day'
    else:
        return 'Regularly Busy Day'
df['busiest_days'] = df.apply(profiles, axis = 1)
```

```python
df.head()
```

```python
freq = df['order_hour_of_day'].value_counts()
print(freq)
```

```python
# adding column from order_hour_of_day to show when the most, average, fewest orders appear.
def profiles(row):
    if row['order_hour_of_day'] in [10,11,14,15,13,12,16,9]:
        return 'Most Orders'
    elif row['order_hour_of_day'] in [17,8,18,19,20,7,21,22]:
        return 'Average Orders'
    else:
        return 'Fewest Orders'
```

```python
df['busiest_period_of_day'] = df.apply(profiles, axis = 1)
```

```python
df.head()
```

```python
df['busiest_period_of_day'].value_counts(dropna = False)
```

```python
# Trying some groupby coding
df.groupby('product_name')
```

```python
df.groupby('department_id').agg({'order_number' : ['mean']})
```

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
# adding a column for max number of orders
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df.head()
```

```python
# adding column for Customer Loyalty
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df['loyalty_flag'].value_counts(dropna = False)
```

```python
# showing statistics for loyalty flag
df.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
#creating a column for average prices by user id
df['avg_prices'] = df.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df.head()
```

```python
# creating flags in new column spender
df.loc[df['avg_prices'] < 10, 'spender'] = 'Low spender'
```

```python
df.loc[df['avg_prices'] >= 10, 'spender'] = 'High spender'
```

```python
df.head()
```

```python
df['spender'].value_counts(dropna = False)
```

```python
# creating new column for average frequency of order
df['avg_freq'] = df.groupby(['user_id'])['days_since_prior_order'].transform(np.mean)
```

```python
df.head()
```

```python
# creating order frequency flags from average frequency column
df.loc[df['avg_freq'] > 20, 'order_freq'] = 'Non-frequent Customer'
```

```python
df.loc[(df['avg_freq'] >10) & (df['avg_freq'] <=20), 'order_freq'] = 'Reguler Customer'
```

```python
df.loc[df['avg_freq'] <=10, 'order_freq'] = 'Frequent Customer'
```

```python
df['order_freq'].value_counts(dropna = False)
```

```python
df.to_pickle(join(path, 'Clean Data', 'final_dataset.pkl'))
```
