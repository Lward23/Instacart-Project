```python
import pandas as pd
import numpy as np
import os
from os.path import join
import pickle
```

```python
path = '../Data'
```

```python
df_pro = pd.read_csv(join(path, 'Clean Data', 'products_clean'))
```

```python
df_main = pd.read_pickle(join(path, 'Clean Data', 'orders_products_prior_merge.pkl'))
```

```python
df_pro.head()
```

```python
# drop Unnamed: 0 column
df_pro = df_pro.drop(columns = ['Unnamed: 0'])
```

```python
df_pro.head()
```

```python
df_main.head()
```

```python
df_pro.shape
```

```python
df_main.shape
```

```python
# merging the main data set with products
df_main2 = df_main.merge(df_pro, on = 'product_id', how = 'inner', indicator = True)
```

```python
df_main2.head()
```

```python
# checking merge results
value_counts = df_main2['_merge'].value_counts()
print(value_counts)
```

```python
# Checking for duplicate values
duplicate_rows = df_main2[df_main2.duplicated()]
if duplicate_rows.shape[0]==0:
    print('No Duplicates Found')
else:
    print(duplicate_rows.head())
```

```python
# checking for missing values
missing_values = df_main2.isnull().sum()
print(missing_values)
```

```python
#dropping merge column
df_main2 = df_main2.drop(columns = ['_merge'])
```

```python
df_main2.head()
```

```python
del df_pro
```

```python
del df_main
```

```python
# saving main2 data set
df_main2.to_pickle(join(path, 'Clean Data', 'orders_prior_products_combined.pkl'))
```

```python
# Next Data set to merge: department
df_dep = pd.read_csv(join(path, 'Clean Data', 'department_clean.csv'))
```

```python
df_dep.head()
```

```python
df_dep = df_dep.rename(columns = {'Unnamed: 0': 'department_id'})
```

```python
df_dep.head()
```

```python
# merging departments with 2nd main dataset
df_main3 = df_main2.merge(df_dep, on = 'department_id', how = 'inner', indicator = True)
```

```python
df_main3.head()
```

```python
# checking main3 results
value_counts = df_main3['_merge'].value_counts()
print(value_counts)
```

```python
# Checking for duplicate values
duplicate_rows = df_main3[df_main3.duplicated()]
if duplicate_rows.shape[0]==0:
    print('No Duplicates Found')
else:
    print(duplicate_rows.head())
```

```python
# checking for missing values
missing_values = df_main3.isnull().sum()
print(missing_values)
```

```python
# dropping the merge column
df_main3 = df_main3.drop(columns = ['_merge'])
```

```python
df_main3.head()
```

```python
# saving main3 data set
df_main3.to_pickle(join(path, 'Clean Data', 'orders_prior_prods_dep.pkl'))
```
